**Unemployment analysis with python**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly .express as px
%matplotlib inline
```

Importing Data

```python
data = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimted Labour Participation Rate','Region','Longitude','Latitude']
```

```python
data.head(267)
```

Checking the correlation

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,6))

ax=plt.gca()
ax.plot(size=10,color='w',labelsize=10,labelcolor='w')

ax= sns.heatmap(data.corr(), annot=True, linewidth=3)

plt.show()
```

Estimated employee in different region

```python
data.columns=['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','Longitude','Latitude']
plt.title('Indian Unemployement')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

Estimated unemployee in different region

```python
plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

Unemployement rate of each indian state

```python
unemployment = data[['States','Region','Estimated Unemployment Rate']]
fig = px.sunburst(unemployment,path=['Region','States'],
                  values='Estimated Unemployment Rate', width=700, height=600,
                  color_continuous_scale='RdY1Gn',title='Unemployment Rate in India')
fig.show()
```
